Loading XML using DOM

```python
import requests
import csv
from xml.dom.minidom import parseString
```

Load XML using request

```python
url = "http://api.irishrail.ie/realtime/realtime.asmx/getCurrentTrainsXML"
page = requests.get(url) # Fetch XML data from the URL
doc = parseString(page.content) # Parse XML content
# check if it works
print(doc.toprettyxml())
```

Write to file

```python
# write to file
with open("trainsxml.xml","w", encoding="utf-8") as f:
    doc.writexml(f)
```

```python
objectTrainPositionsNode = doc.getElementsByTagName("objTrainPositions")
import csv
with open("trains.csv", "w", newline='', encoding="utf-8") as fh:
    csv_writer = csv.writer(fh)
    for objTrainPositionsNode in objectTrainPositionsNode:
        trainStatus = objTrainPositionsNode.getElementsByTagName("TrainDate")[0].firstChild.nodeValue.strip()
        trainLatitude = objTrainPositionsNode.getElementsByTagName("TrainLatitude")[0].firstChild.nodeValue.strip()
        trainLongitude = objTrainPositionsNode.getElementsByTagName("TrainLongitude")[0].firstChild.nodeValue.strip()
        trainDate = objTrainPositionsNode.getElementsByTagName("TrainDate")[0].firstChild.nodeValue.strip()
        publicMessage = objTrainPositionsNode.getElementsByTagName("PublicMessage")[0].firstChild.nodeValue.strip()
        direction = objTrainPositionsNode.getElementsByTagName("Direction")[0].firstChild.nodeValue.strip()
        csv_writer.writerow([trainStatus,trainLatitude,trainLongitude,trainDate,publicMessage,direction])

    
```

```python
retrieveTags=['TrainStatus',
'TrainLatitude',
'TrainLongitude',
'TrainCode',
'TrainDate',
'PublicMessage',
'Direction']
with open("trains.csv", "w", newline='', encoding="utf-8") as fh:
    csv_writer = csv.writer(fh)
    csv_writer.writerow(retrieveTags)  # Write header row
    for objTrainPositionsNode in objectTrainPositionsNode:
        row = [objTrainPositionsNode.getElementsByTagName(tag)[0].firstChild.nodeValue.strip()   for tag in retrieveTags]
        csv_writer.writerow(row)
```
